## Generates a 60Hz sine wave from 0 to 255 (value levels)

### ploting the wave

```python
import matplotlib.pyplot as plt
import numpy as np

f = 60.0	# Hz
v_min = 0   # steps
v_max = 255 # steps

samples = 100

t = np.linspace(0, 1/f, samples+1)
y = np.uint64((np.sin(2*np.pi*f*t) + 1) * (v_max-v_min) / 2)

plt.plot(t, y)	
plt.title('Sin wave (60Hz)')
plt.xlabel('time [s]')
plt.ylabel('amplitude [steps]')
plt.show()
```

### generate the table to copy on Aixt

```python
print('const y = [\n\tu8({}), '.format(y[0]), end='')
for i in range(1, len(y)-1): 
	if (i % 10) == 0 :
		print('\n\t', end='')
	print('{}, '.format(y[i]), end='')
print('\n]')
```

### sampling frequency

```python
print('// sampling frequency: {}'.format((1/f)/samples))
```
